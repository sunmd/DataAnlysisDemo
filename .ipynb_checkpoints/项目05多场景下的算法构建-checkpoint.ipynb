```python
'''
【项目05】  多场景下的算法构建

课程数据：
某公司A,B产品在2018年1,2,3月的销量数据，数据格式为xlsx

作业要求：
1、批量读取数据，并输出以下信息
（1）数据量
（2）数据字段columns
（3）输出每个文件分别有多少缺失值
要求：
① 创建独立函数，从读取数据到以上输出要求
② 运行代码多次调用创建函数，对数据进行批量处理
提示：
① 将课程excel数据放入单独文件夹(建议英文路径)，函数中通过input来输入excel数据所在文件夹路径，再通过遍历来读取文件
② pd.read_excel()中用“index_col”参数，将第一列变为index
③ os.walk(path) → 返回路径中的信息及文件，结果为一个生成器
④ 定位缺失值位置：data[data.isnull().values == True]

2、批量读取数据，用均值填充缺失值数据，并完成以下计算及图表制作
（1）读取数据并用均值填充缺失值；对“日期”字段进行时间序列处理，转换成日period ，最后输出三个Dataframe文件data1,data2,data3
（2）分别计算data1，data2，data3中A,B产品的月总销量，并绘制多系列柱状图，存储在对应的图片文件夹路径
（3）分别计算A产品在每个月中哪一天超过了月度80%的销量，输出日期
要求：
① 分别创建四个函数，对应完成上述需求
② 数据文件夹路径和图片存储路径不要相同
提示：
① 时间序列转换方法提示：pd.to_period
② 时间戳Timestampe转化成字符串用str()方法

3、读取数据并合并，做散点图观察A,B产品销量，并做回归，预测当A销量为1200时，B产品销量值
（1）读取数据删除缺失值；对“日期”字段进行时间序列处理，转换成日period ，合并三个月数据，输出data；
（2）针对A产品销量和B产品销量数据做回归分析，制作散点图并存储，并预测当A销量为1200时，B产品销量值
要求：
① 分别创建两个函数，对应完成上述需求
② 数据文件夹路径和图片存储路径不要相同
提示：
① 用pd.dripna方法去掉缺失值，注意inplace参数

'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['KaiTi']
mpl.rcParams['font.serif'] = ['KaiTi']
mpl.rcParams['axes.unicode_minus']=False


% matplotlib inline
```

```python
'''
作业要求：
1、批量读取数据，并输出以下信息
（1）数据量
（2）数据字段columns
（3）输出每个文件分别有多少缺失值
要求：
① 创建独立函数，从读取数据到以上输出要求
② 运行代码多次调用创建函数，对数据进行批量处理
提示：
① 将课程excel数据放入单独文件夹(建议英文路径)，函数中通过input来输入excel数据所在文件夹路径，再通过遍历来读取文件
② pd.read_excel()中用“index_col”参数，将第一列变为index
③ os.walk(path) → 返回路径中的信息及文件，结果为一个生成器
④ 定位缺失值位置：data[data.isnull().values == True]
'''
#/home/sunmd/文档/python/【非常重要】Python数据分析师微专业_项目资料/项目05多场景下的算法构建
#读取路径的函数，进行文件扫描，然后读取数据
def read_file():
    
    lst = []
    
    file_path = input("请输入excel数据所在文件路径: \n")
    gen = os.walk(file_path, topdown=True)
    
    #进行文件数据扫描
    for root, dirs, files in gen:
        
        #读取文件数量
        file_num = 1
        for obj_file in files:
            #查看是否有xlsl后缀的文件
            if '.xlsx' in obj_file:
                
                print('搜索到对应文件：%s' %(root+'/'+obj_file))
                temp_df = read_data(root+'/'+obj_file)
                print('第%d个数据数据量为：%d' %(file_num, len(temp_df)))
                print('第%d个数据数据字段为：%s' %(file_num, temp_df.columns))
                print('第%d个数据缺失值数量为：%d' %(file_num, len(temp_df[temp_df.isnull().values == True])))
                print('-'*10)
                lst.append(temp_df)
    return lst

#提供文件路径然后读取对应的数据文件
def read_data(file_path):
#     temp = '/home/sunmd/文档/python/【非常重要】Python数据分析师微专业_项目资料/项目05多场景下的算法构建/data01.xlsx'
    df = pd.read_excel(file_path, index_col=0)
    return df

lst_bak = read_file()
```

```python
'''
2、批量读取数据，用均值填充缺失值数据，并完成以下计算及图表制作
（1）读取数据并用均值填充缺失值；对“日期”字段进行时间序列处理，转换成日period ，最后输出三个Dataframe文件data1,data2,data3
（2）分别计算data1，data2，data3中A,B产品的月总销量，并绘制多系列柱状图，存储在对应的图片文件夹路径
（3）分别计算A产品在每个月中哪一天超过了月度80%的销量，输出日期
要求：
① 分别创建四个函数，对应完成上述需求
② 数据文件夹路径和图片存储路径不要相同
提示：
① 时间序列转换方法提示：pd.to_period
② 时间戳Timestampe转化成字符串用str()方法
'''
#返回data1/data2/data3，用均值填充缺失值
def f2_1(lst):
    data1 = lst[0]
    data2 = lst[2]
    data3 = lst[1]
    for temp_data in data1,data2,data3:
        
        #读取数据并用均值填充缺失值
        for col in temp_data.columns:
            temp_data[col].fillna(temp_data[col].mean(), inplace=True)
            
        #转换成日period
        temp_data['月份'] = temp_data.index.to_period('M')

        
    print('运行函数f2_1()，分别得到data1,data2,data3')
    print('-'*50)
    return data1, data2, data3

lst1 = lst_bak.copy()

data1,data2,data3 = f2_1(lst1)
print(data1.index.dtype)

#分别计算data1，data2，data3中A,B产品的月总销量，并绘制多系列柱状图，存储在对应的图片文件夹路径
def f2_2(save_path):
    
    #创建空表，用于统计所有数据
    total_data = pd.DataFrame(columns=data1.columns)

    for temp_data in data1,data2,data3:
        total_data = total_data.append(temp_data)
    print(total_data.index.dtype)
    total_data = total_data.groupby('月份').sum()
    print(total_data)
    print('运行函数f2_2()，分别计算data1，data2，data3中A,B产品的月总销量，并绘制多系列柱状图，存储在对应的图片文件夹路径')
    print('-'*50)
    
    #做图
    fig = plt.figure(num=1, figsize=(16,10))
    total_data.plot(kind='bar', grid=True, alpha=0.8, rot=0)
    plt.savefig(save_path, dpi=400)
    
#保存图片路径
save_path = '/home/sunmd/文档/python/total.png'

f2_2(save_path)

def f2_3():
    
    temp_lst = []
    for temp_data in lst_bak:
        temp_data['productA_nor'] = temp_data['productA']/temp_data['productA'].sum()
        #标准化，累积
        temp_data['productA_nor_cu'] = temp_data['productA_nor'].cumsum()
        
        #过滤出来超过0.80的那一天
        temp_lst.append(str(temp_data[temp_data['productA_nor_cu']>0.80].iloc[0].name))

        
    print(temp_lst)
    print('分别计算A产品在每个月中哪一天超过了月度80%的销量，输出日期')
    print('-'*50)
    return temp_lst

f2_3()  
```

```python
'''
3、读取数据并合并，做散点图观察A,B产品销量，并做回归，预测当A销量为1200时，B产品销量值
（1）读取数据删除缺失值；对“日期”字段进行时间序列处理，转换成日period ，合并三个月数据，输出data；
（2）针对A产品销量和B产品销量数据做回归分析，制作散点图并存储，并预测当A销量为1200时，B产品销量值
要求：
① 分别创建两个函数，对应完成上述需求
② 数据文件夹路径和图片存储路径不要相同
提示：
① 用pd.dripna方法去掉缺失值，注意inplace参数
'''
def f3_1():
    lst = []
    
    file_path = '/home/sunmd/文档/python/【非常重要】Python数据分析师微专业_项目资料/项目05多场景下的算法构建'
    gen = os.walk(file_path, topdown=True)
    
    #进行文件数据扫描
    for root, dirs, files in gen:
        
        #读取文件数量
        file_num = 1
        for obj_file in files:
            #查看是否有xlsl后缀的文件
            if '.xlsx' in obj_file:
                
                temp_df = read_data(root+'/'+obj_file)
                lst.append(temp_df)
                
    total_df = pd.DataFrame(columns=lst[0].columns)

    #合并所有数据
    for obj_df in lst:
        total_df = total_df.append(obj_df)
        
    #安索引排序 
    total_df.sort_index(inplace=True)
    
    #用pd.dripna方法去掉缺失值，注意inplace参数
    total_df.dropna(axis=0, inplace=True)
    
    #对“日期”字段进行时间序列处理，转换成日period
    total_df.index = total_df.index.to_period('D')

    return total_df
    
def f3_2(demo_data, test):
    
    xtrain = demo_data['productA']
    ytrain = demo_data['productB']
    
    model = LinearRegression()
    model.fit(xtrain[:,np.newaxis], ytrain)
    
    xtest = test
    ytest = model.predict(xtest[:,np.newaxis])
    
    print("特征因子：%f"%model.coef_)
    print("偏移量  ：%f"%model.intercept_)
    
    return xtest,ytest
    

data = f3_1()

xtest = np.linspace(0, 1000, 1000)

test_data = f3_2(data, xtest)

#做散点图
fig = plt.figure(num=0, figsize=(16,12))
plt.scatter(data['productA'], data['productB'], alpha=0.8)
plt.grid(True)
plt.xlim([-200,1200])
plt.ylim([0,1400])
plt.title('A-B产品销量回归拟合')

#预测回归线
plt.scatter(test_data[0], test_data[1], marker='.', color='r')

#保存图片
plt.savefig('/home/sunmd/文档/python/hg.png', dpi=400)

yc = f3_2(data,np.array([1200]))

print("预测当产品A销量为%d，产品B销量为%.1f" %(yc[0], yc[1]))
```
